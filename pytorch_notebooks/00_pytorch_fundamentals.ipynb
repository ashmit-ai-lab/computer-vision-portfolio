### Random tensors

## Reshaping, stacking, squeezing and unsqueezing tensors
-> Reshaping - reshapes an input tensor to a defined shape.

-> View - Return a view of an input tensor of certain shape but
          keep the same memory as the original tensor.

-> Stacking - Combine multiple tensors on top of each other(vstack) or side by side(hstack).

-> Squeeze - Removes all '1' dimensions from a tensor.

-> Unsqueeze - Add a '1' dimension to a target tensor.

-> Permute - Return a view of the input with dimensions permuted(swapped) in a certain way.


```python
#Let's create a tensor
import torch
x = torch.arange(1,13)
x, x.shape
```

```python
#Add an extra dimension
x_reshaped = x.reshape(3,4)
x_reshaped, x_reshaped.shape
```

```python
#Change the view
z = x.view(1,12)
z,z.shape, x, x.shape
```

```python
#Changing z changes x
 # (because a view of a tensor shares the same memory as the original input)
z[:, 0] = 5
z,x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x,x,x,x], dim = 0)
x1_stacked = torch.stack([x,x,x,x], dim = 1)
x_stacked, x1_stacked
```

```python
# torch.squeeze() - removes all single dimensions from a target tensor
x_reshaped = x_reshaped.reshape(1,12)
x_reshaped, x_reshaped.shape
```

```python
# torch.squeeze() - removes all single dimensions from a target tensor
print(f"Previous tensor: {x_reshaped}")
print(f"Previous tensor shape: {x_reshaped.shape}")

#Remove extra dimensions from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New tensor shape: {x_squeezed.shape}")
```

```python
# torch.unsqueeze() - Adds a single dimension to a target tensor at a specific dimension
print(f"\nPrevious tensor: {x_squeezed}")
print(f"Previous tensor shape: {x_squeezed.shape}")

#Adds extra dimension to x_squeezed
x_unsqueezed = x_squeezed.unsqueeze(dim = 0)
x_unsqueezed1 = x_squeezed.unsqueeze(dim = 1)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New tensor shape: {x_unsqueezed.shape}")

#print(f"\nNew tensor: {x_unsqueezed1}")
#print(f"New tensor shape: {x_unsqueezed1.shape}")
```

```python
#torch.permute rearranges the dimensions of a target tensor in a specified order
x_org = torch.randn(224,224,3) # [height, width, colour_channels] of an image

#Permute the original tensor to rearrange the axis (or dim) order
#[cc, h, w]
#Permute shares the same memory as x original
x_permuted = torch.permute(x_org,(2,0,1)) #shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_org.shape}")
print(f"New shape: {x_permuted.shape}")
```

```python
print(f"Previous value: {x_org[0, 0, 0]}")
x_org[0, 0, 0] = 322
print(f"New value: {x_permuted[0, 0, 0]}")
```

## Reproducibility (trying to take random out of random)

In short how a neural network learns is

* start with random numbers -> tensor operations -> update them to make them better representations of the data -> repeat

* To reduce the randomness in NNs and pytorch comes the concept of a **random seed**

Essentially what the random seed does is "flavour" the randomness. (Pseudorandomness)

Extra resources - https://docs.pytorch.org/docs/stable/notes/randomness.html

Wiki - https://en.wikipedia.org/wiki/Random_seed

```python
torch.rand(3,3) #every time new numbers
```

```python
import torch

# Create two random tensors
random_tensor_a = torch.rand(3,4)
random_tensor_b = torch.rand(3,4)

print(random_tensor_a)
print(random_tensor_b)
print(random_tensor_a == random_tensor_b)
```

```python
# Let's make some random but reproducible tensors
import torch

# Set the random seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_c = torch.rand(3,4)
torch.manual_seed(RANDOM_SEED)
random_tensor_d = torch.rand(3,4)

print(random_tensor_c)
print(random_tensor_d)
print(random_tensor_c == random_tensor_d)
```
